```python
import pandas as pd
import plotly.express as px

car_data = pd.read_csv('../vehicles_us.csv') # leer los datos

# Histograma del odómetro con color fijo
fig = px.histogram(car_data, x="odometer", 
                   title="Distribución del Odómetro",
                   color_discrete_sequence=['blue'])
fig.show()

# Histograma de precios con color fijo
fig = px.histogram(car_data, x="price", 
                   title="Distribución de Precios",
                   color_discrete_sequence=['red'])
fig.show()
```

```python
fig = px.scatter(car_data, x="odometer", y="price",
                 title="Relación entre Kilometraje y Precio",
                 labels={'odometer': 'Kilometraje', 'price': 'Precio ($)'})
fig.show()

# Colorear por tipo de vehículo
fig = px.scatter(car_data, x="odometer", y="price", 
                 color="type",
                 title="Precio vs Kilometraje por Tipo de Vehículo")
fig.show()

# O colorear por condición
fig = px.scatter(car_data, x="odometer", y="price", 
                 color="condition",
                 title="Precio vs Kilometraje por Condición")
fig.show()
```
